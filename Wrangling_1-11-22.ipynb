```python
genres_df = pd.read_csv("genre_wrangling.csv")
```

```python
df['track_artists_genres'] = df['track_artists_genres'].apply(lambda v: eval(v))
```

```python
genres_dict = {}
```

```python
for index in genres_df.index:
    genres_dict[genres_df.iloc[index,1]] = genres_df.iloc[index,2]
```

```python
track_artists_genres_list = list(df["track_artists_genres"])
```

```python
reshaped_list = [original_list[0] if type(original_list[0]) == list else original_list for original_list in track_artists_genres_list]
```

```python
df["track_artists_genres"] = reshaped_list
```

```python
df
```

```python
df["track_artists_genres_consolidate"] = df["track_artists_genres"].apply(lambda l: [genres_dict.get(v) for v in l])
```

```python
df["track_artists_genres_consolidate"]
```

```python
def dummy_code(l, new_col):
    if new_col in l:
        return(1)
    else:
        return(0)
```

```python
for new_col in genres_df['New Genre'].unique():
    df[new_col] = df["track_artists_genres_consolidate"].apply(lambda v: dummy_code(v, new_col))
```

```python
df['track_artists_popularity'] = df['track_artists_popularity'].apply(lambda v: eval(v))
```

```python
df['track_artists_followers'] = df['track_artists_followers'].apply(lambda v: eval(v))
```

```python
def list_recode(v):
    if type(v) == list:
        return(v[0])
    else:
        return(v)
```

```python
df['track_artists_popularity_first'] = df['track_artists_popularity'].apply(lambda v: list_recode(v))
```

```python
df['track_artists_followers_first'] = df['track_artists_followers'].apply(lambda v: list_recode(v))
```

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
df.to_csv("allchartalbumtracks_RECODED.csv")
```

```python
df = pd.read_csv("allchartalbumtracks_RECODED.csv",index_col=False)
```

```python
df['track_artists_followers'] = df['track_artists_followers'].apply(lambda v: eval(v))
```

```python
def recode(v):
    if type(v) == list:
        return(v)
    else:
        return([int(v)])
```

```python
df['track_artists_followers'] = df['track_artists_followers'].apply(lambda v: recode(v))
```

```python
df['track_artists_combined_followers'] = df['track_artists_followers'].apply(lambda v: sum(np.array(v)))
```

```python
df.to_csv("allchartalbumtracks_RECODED_1-11.csv")
```
